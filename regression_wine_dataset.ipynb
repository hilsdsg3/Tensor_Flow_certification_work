# import libraries

```python
import pandas as pd
import csv
import requests
from io import BytesIO
import zipfile
import os
from sklearn.model_selection import train_test_split
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Pre-processing the data

```python
red_wine = pd.read_csv(path + '\\'+ 'winequality-red.csv', sep=';')
white_wine = pd.read_csv(path + '\\'+ 'winequality-white.csv', sep=';')

print(f'\nTotal red wines in this dataset : {len(red_wine)}')
print(display(red_wine.sample(5)))
print(f'\nTotal white wines in this dataset : {len(white_wine)}')
print(display(white_wine.sample(5)))
```

## Creation of the variables

```python
wine = red_wine.append(white_wine)
wine = wine.reset_index(drop=True)
print(display(wine.sample(10)))

X = wine.drop('quality', axis=1)
print(X.columns)

y = wine['quality']
```

# split variables

```python
# Create training and test sets
train_split = 0.8
test_split = 1 - train_split  

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=test_split, 
                                                    random_state=42) # set random state for reproducible splits
```

```python
print(f'X train : {X_train.shape}')
print(f'X test : {X_test.shape}')
print(f'Y train : {y_train.shape}')
print(f'Y test : {y_test.shape}')
```

```python

def model_parameters(seed_,
                     units_for_layers,
                     optimizer_,
                     loss_metric,
                     model_summary,
                     epochs_,
                     verbose_,
                     X_train_,
                     y_train_,
                     X_test_,
                     compiled_model_view,
                     predict_,
                     evaluate_,
                     loss_epoch_plot):
    
    if optimizer_ == 'Adam':
        optimizer_ = tf.keras.optimizers.Adam()
    elif optimizer_ == 'SGD':
        optimizer_ = tf.keras.optimizers.SGD()
        
    # Set random seed
    tf.random.set_seed(seed_)
    
    # loss metric
    if loss_metric == 'mae':
        loss_ = tf.keras.losses.mae
    elif loss_metric == 'mse':
        loss_ = tf.keras.losses.mse

    # Create a new model (same as model_2)
    model_start_ = 'tf.keras.Sequential(['
    model_end_ = '])'
    model_middle = ''
    
    for nr,i in enumerate(units_for_layers):
        if nr < 1:
            temp = 'tf.keras.layers.Dense(' + str(i) + ')'
            model_middle = temp + model_middle 
        else:    
            temp = 'tf.keras.layers.Dense(' + str(i) + ')'
            model_middle = model_middle + ', ' + temp
            
    # Create a model using the sequntial API        
    model_ = eval(model_start_ + model_middle + model_end_)

    # Compile the model
    compiled_model = model_.compile(loss=loss_,
                                    optimizer=optimizer_,
                                    metrics=[loss_metric])
    if compiled_model_view:
        print(compiled_model)    
        
    # Fit the model
    fit_model = model_.fit(X_train_, y_train_, epochs=epochs_, verbose=verbose_)

    # Print model summary
    if model_summary:
        print(f'MODEL SUMMARY :')
        print(model_.summary())
        
    # Evaluate
    if evaluate_:
        print(f'\nEVALUATE (VALIDATION) :')
        print(model_.evaluate(X_test, y_test))
    

    # loss_epoch_plot
    if loss_epoch_plot:
        print('\n')
        pd.DataFrame(fit_model.history).plot()
        plt.ylabel('loss')
        plt.xlabel('epochs')
    
    
    return(fit_model)
    
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=y_pred):
    '''
    Plots training data, test data and compares prediction.
    '''
    plt.figure(figsize=(10, 7))
    # Plot training data in blue
    plt.scatter(train_data, train_labels, c="b", label="Training data")
    # Plot training data in green
    plt.scatter(test_data, test_labels, c="g", label="Testing data")
    # Plot training data in green
    plt.scatter(test_data, predictions, c="r", label="Predictions")
    # Show the legend
    plt.legend();
```
